# Section 1: Data Preprocessing

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline

directory = './stock_data/' 
df = pd.read_csv('updated_companylist.csv', header= None)
list_of_stocks = df[0].tolist()
```

```python
d = {} #489
for i in list_of_stocks:
    data = pd.read_csv(directory+i+'.csv', index_col = 'Date', parse_dates = ['Date'])
    data.drop(['High', 'Low', 'Open', 'Volume', 'Adj Close'], axis=1, inplace = True)
    d[i] = data
len(d)
```

```python
d['AABA'].head()
```

### Split train-test data set

```python
new_d = {} #489

for i in list_of_stocks:
    length = len(d[i])
    new_d[i] = {}
    training_size = int(0.75*length)
    new_d[i]['Train'] = d[i][:training_size]
    new_d[i]['Test'] = d[i][training_size:]
```

```python
new_d['AABA']['Test'].head()
```

```python
i = 'AABA'
plt.figure(figsize=(14,4))
plt.plot(new_d[i]["Train"])
plt.plot(new_d[i]["Test"])
plt.title(i + " Closing Stock Price")
plt.legend(["Training Set", "Test Set"])
plt.xlabel("Date")
plt.ylabel("Price")
```

### Normalize data

```python
transform_train = {}
transform_test = {}
scalerset = {}
for i in list_of_stocks:
    scaler = MinMaxScaler(feature_range=(0,1))
    train = np.array(new_d[i]["Train"])
    test = np.array(new_d[i]["Test"])
    train = train.reshape(train.shape[0],1)
    test = test.reshape(test.shape[0],1)
    transform_train[i] = scaler.fit_transform(train) # Scaler not stored as not required to scale back train_set 
    transform_test[i] = scaler.fit_transform(test) # Scaler to be stored to scale back test_set for plotting predicted values
    scalerset[i] = scaler
```

```python
# print(transform_train['AABA'])
# print(transform_test['AABA'])
```

### Set window size

```python
win_size = 20
train_set = {}
test_set = {}

for i in list_of_stocks:
    train_set[i] = {}
    X_train = []
    y_train = []
    for j in range(win_size, len(new_d[i]['Train'])):
        X_train.append(transform_train[i][j-win_size:j,0])
        y_train.append(transform_train[i][j,0])
    X_train, y_train = np.array(X_train), np.array(y_train)
    train_set[i]["X"] = np.reshape(X_train, (X_train.shape[0],X_train.shape[1],1))
    train_set[i]["y"] = y_train
    
    test_set[i] = {}
    X_test = []
    y_test = []    
    for j in range(win_size, len(new_d[i]['Test'])):
        X_test.append(transform_test[i][j-win_size:j,0])
        y_test.append(transform_test[i][j,0])
    X_test, y_test = np.array(X_test), np.array(y_test)
    test_set[i]["X"] = np.reshape(X_test, (X_test.shape[0], X_train.shape[1], 1))
    test_set[i]["y"] = y_test
```

# Section 2: Build the LSTM model

```python
from keras.models import Sequential
from keras.optimizers import RMSprop
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
np.random.seed(42)
```

```python
%%time
# The LSTM model
model = Sequential()
# 1st input layer with dropout regularisation
model.add(LSTM(units=win_size, input_shape=(X_train.shape[1],1), return_sequences=True))
model.add(Dropout(0.2))
# 2nd input layer
model.add(LSTM(units=win_size, return_sequences=True))
model.add(Dropout(0.2))
# 3rd input layer
model.add(LSTM(units=win_size, return_sequences=True))
model.add(Dropout(0.5))
# 4th input layer
model.add(LSTM(units=win_size))
model.add(Dropout(0.5))
# output layer
model.add(Dense(units=1))

RMSprop(lr=0.1, rho=0.9, epsilon=None, decay=0.0)

# compile model
model.compile(optimizer='rmsprop', loss='mean_squared_error')
# fit model with train_set
for i in list_of_stocks:
    print(i)
    model.fit(train_set[i]["X"], train_set[i]["y"], batch_size=300, epochs=5)
```

# Section 3: Save the trained model

```python
import os
# create folder to store saved models
if((os.path.exists('saved_models'))==False):
    os.mkdir('./saved_models')
    
# save model
model.save('./saved_models/final_model.h5')  # creates a HDF5 file (.h5)
```

# Section 5: Predict future stock prices

### Predict using Test Set

```python
prediction_result = {}
for i in list_of_stocks:
    predicted = scalerset[i].inverse_transform(model.predict(test_set[i]["X"]))
    actual = scalerset[i].inverse_transform(test_set[i]["y"].reshape(-1,1))
    MSE = mean_squared_error(actual, predicted)
    prediction_result[i] = {}
    prediction_result[i]["Pred"] = predicted
    prediction_result[i]["True"] = actual
    prediction_result[i]['MSE'] = MSE
```

```python
test_date = new_d['AAPL']['Test'].reset_index().iloc[20:]['Date']
len(test_date)
```

```python
# plot AAPL
plt.figure(figsize=(14,6))
plt.plot(test_date, prediction_result['AAPL']['True'])
plt.plot(test_date, prediction_result['AAPL']['Pred'])
plt.title("{} with MSE {:10.4f}".format('AAPL',prediction_result['AAPL']['MSE']))
plt.legend(["True", "Pred"])
plt.xlabel("Date")
plt.ylabel("Price")
```

```python
# plot MSFT
plt.figure(figsize=(14,6))
plt.plot(test_date, prediction_result['MSFT']['True'])
plt.plot(test_date, prediction_result['MSFT']['Pred'])
plt.title("{} with MSE {:10.4f}".format('MSFT',prediction_result['MSFT']['MSE']))
plt.legend(["True", "Pred"])
plt.xlabel("Date")
plt.ylabel("Price")
```

```python
import math

mse_sum = 0.0
rmse_sum = 0.0

for i in list_of_stocks:
    mse_sum = mse_sum + prediction_result[i]['MSE']
    rmse_sum = rmse_sum + math.sqrt(prediction_result[i]['MSE'])

print('avg_mse:' + str(mse_sum/len(list_of_stocks)))
print('avg_rmse:' + str(rmse_sum/len(list_of_stocks)))
```

```python
# set vector with window size
X_unseen_test = []
y_unseen_test = []    
for j in range(win_size, len(unseen_data)):
    X_unseen_test.append(unseen_transform_test[j-win_size:j,0])
    y_unseen_test.append(unseen_transform_test[j,0])
X_unseen_test, y_unseen_test = np.array(X_unseen_test), np.array(y_unseen_test)
unseen_testset = np.reshape(X_unseen_test, (X_unseen_test.shape[0], win_size, 1))
# test_set = y_test
```
